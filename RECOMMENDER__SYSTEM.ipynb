```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
def stem(text):
    y=[]
    ps = PorterStemmer()
  
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```
